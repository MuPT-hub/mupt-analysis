```python
import MDAnalysis as mda
from MDAnalysis.analysis.rdf import InterRDF
import matplotlib.pyplot as plt
u = mda.Universe("/content/drive/Shareddrives/MuPT Project Coordination/MuPT Analysis/MDA_feature_code_using_python/production_4ns.dump", format="LAMMPSDUMP")
u.trajectory[0]

# Print atom types or names to help you choose CG beads
print(u.atoms.types)

# print(u.atoms.names)

group1 = u.select_atoms("type 1 and index 1")
group2 = u.select_atoms("type 1")

```

```python
group1
```

```python
# For RDF calculation (defaults: bins=75, range=(0, 15))
rdf = InterRDF(group1, group2, nbins=200, range=(0.0, 20.0))
rdf.run(verbose=True)

# Plot RDF
plt.figure(figsize=(8, 5))
plt.plot(rdf.bins, rdf.rdf, label='RDF (type 1 vs type 1)')
plt.xlabel('Distance (Å)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# For RDF calculation (defaults: bins=75, range=(0, 15))
rdf = InterRDF(group1, group2, nbins=100, range=(0.0, 20.0))
rdf.run(verbose=True)

# Plot RDF
plt.figure(figsize=(8, 5))
plt.plot(rdf.bins, rdf.rdf, label='RDF (type 1 vs type 1)')
plt.xlabel('Distance (Å)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
